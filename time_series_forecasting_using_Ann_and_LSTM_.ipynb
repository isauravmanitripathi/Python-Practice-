```python
import numpy as np 
import pandas as pd 
import plotly.express as px 
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras import Sequential , layers
from tensorflow.keras.metrics import RootMeanSquaredError
from sklearn.metrics import mean_squared_error, r2_score
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./Data/1979-2021.csv", parse_dates = ['Date'])
df
```

```python
df_india = df[['Date', 'India(INR)']].set_index('Date')
df_india
```

```python
fig = px.line(df_india,title='Monthly Average Gold Price, India',labels={'Date':'Months','value':'Average Gold Price'})
fig.update_layout(title={'font_family':'Georgia','font_size':23,'x':0.5})
fig.show() 
```

```python
from statsmodels.tsa.stattools import acf
acf_df = pd.DataFrame({'ACF':acf(df_india, nlags = df.shape[0])})
acf_df['Upper Level'] = 1.96 / 1.96 / (df.shape[0] ** 0.5)
acf_df['Lower Level'] = - 1.96 / 1.96 / (df.shape[0] ** 0.5)

fig = px.line(acf_df, title = 'Autocorrelation Plot', labels = {'index':'lags', 'value':'Correlation'})
fig.update_layout(title = {'font_family': 'Georgia', 'font_size':23, 'x':0.5})
fig.show()
```

```python
def lag_plot(lags = 1):
  x = df_india['India(INR)']
  y = df['India(INR)'].shift(lags)
  title = 'Lag Plot of lag' + str(lags)
  fig = px.scatter(x=x,y=y,trendline='ols',trendline_color_override='red', title = title, 
                      labels={'x':'Gold Prices','y':'lag of {} on Gold Prices'.format(lags)})
  fig.update_layout(title={'font_family':'Georgia','font_size':23,'x':0.5})
  fig.show() 
```

```python
lag_plot(1)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decom_df = pd.Series([x%3 for x in range(100)])
fig = px.line(decom_df.trend,title='Trend')
fig.update_layout(title={'font_family':'Georgia','font_size':23,'x':0.5})
fig.show()
fig = px.line(decom_df.seasonal,title='Seasonality')
fig.update_layout(title={'font_family':'Georgia','font_size':23,'x':0.5})
fig.show()
fig = px.line(decom_df.resid,title='Residuals')
fig.update_layout(title={'font_family':'Georgia','font_size':23,'x':0.5})
fig.show()
```
